```python
# 職涯性向測驗

# 定義12個問題，每個問題對應到一個類型
questions = [
    ("你是否喜歡處理數字和邏輯問題？", 0),
    ("你是否注重細節並且善於發現錯誤？", 0),
    ("你是否常常埋頭苦幹，不太喜歡與人交流？", 0),
    ("你是否善於完成任務且腳踏實地？", 1),
    ("你是否刻苦耐勞並且在意時機？", 1),
    ("你是否體貼他人但有時優柔寡斷？", 1),
    ("你是否有很多創意點子並且行動快速？", 2),
    ("你是否自我意識強且果斷？", 2),
    ("你是否樂觀但有時缺乏耐心？", 2),
    ("你是否充滿大愛並且喜歡熱鬧？", 3),
    ("你是否感性且覺得自己超棒？", 3),
    ("你是否愛說話但有時粗線條？", 3)
]

# 計分
scores = {0: 0, 1: 0, 2: 0, 3: 0}

print("歡迎參加職涯性向測驗，請回答以下問題（是Y/否N）:")
for q, t in questions:
    answer = input(q + " ")
    if answer.strip() in ["是", "yes", "Y", "y"]:
        scores[t] += 1

# 找出最高分的類型
max_type = max(scores, key=scores.get)

# 根據類型輸出結果
representatives = {
    0: "股神巴菲特",
    1: "郭台銘",
    2: "伊隆·馬斯克",
    3: "黃仁勳"
}

traits = {
    0: "數字力，邏輯力",
    1: "服務力，複製力",
    2: "創造力，品牌力",
    3: "自帶流量，超級巨星"
}

name = input("請輸入您的姓名：")

print(f"\n{name}，您好")
print(f"根據職涯性向測驗的結果顯示，您是屬於{max_type}類型的人格特質")
print(f"該類型的代表人物有{representatives[max_type]}")
print("選擇職務的傾向，最好是配合自已的性向")
print("如果是要創業的話，一個團隊最好有3種不同性格的人")

# 額外資訊
print("\n同學選一種適合性向工作發展，切勿跨其他性向，除非你當老闆")
print("如果被調動，轉職，切勿做你個性不適合的工作")
print("\n一個團隊最好有3種不同性格的人")
print(f"0的人相對3的人，個性完全不合，")
print(f"2的人相對1的人，個性完全不合，")
print("不太適合婚姻生活")
```
